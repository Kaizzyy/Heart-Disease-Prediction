# Heart Disease Prediction. Collecting Data

#### Import neccessary libraries

```python
import pandas as pd 
```

#### Read data from '*.data' file

```python
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
data_clev = pd.read_csv("processed.cleveland.data", sep=",", names=columns)
data_hung = pd.read_csv("processed.hungarian.data", sep=",", names=columns)
data_swit = pd.read_csv("processed.switzerland.data", sep=",", names=columns)
data_va = pd.read_csv("processed.va.data", sep=",", names=columns)
```

#### Collect four dataset to one dataframe

```python
data_all = pd.concat([data_clev, data_hung, data_swit, data_va])
```
